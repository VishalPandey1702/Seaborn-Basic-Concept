```python
import pandas as pd 
import numpy as np 
import matplotlib as plt
import seaborn as sns
plt.style.use("ggplot") # it wiil give the style to the graph
data = sns.load_dataset("tips") # it is use to load the data
data.head()
```

```python
# plot of total bill vs tip
sns.relplot(x="total_bill",y="tip",data=data) 
```

```python
# it will diferentiate othe basis of the sex in the style
sns.relplot(x="total_bill",y="tip",hue = "smoker",data=data) # it differentiate the smoker and non smoker on the basis of the color
```

```python
sns.relplot(x="total_bill",y="tip",hue = "smoker",style = "sex", data=data)
```

```python
# on  the basis ofthe size (it will take only the numeric data only)
sns.relplot(x="total_bill",y="tip",hue = "smoker",style="sex",size="size",sizes=(11,110),data=data)
# size --> it differentiate on the base of size, sizes --> it will provide the size to the data 
```

```python
# 1 . CATOGORICAL SCATTER PLOT
# Catogorical data --> X-axis
# Numerical data --> Y-axis
tip =sns.load_dataset("tips")
tip.head()
```

```python
# a. strip plot
sns.catplot(x="day",y="tip",kind = "strip",jitter=0.5,data=tip) # jitter --> difference between the points
```

```python
# b. swarm plot (sam point are at the side by side)
sns.catplot(x='day',y='tip',kind='swarm',data=tip)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue="sex",data=tip)
```
